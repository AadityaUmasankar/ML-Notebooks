```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.set_option('display.max_columns',1000)
```

```python
df = pd.read_csv('datasets/data.csv')
```

```python
df.head()
```

```python
df.drop(['car_ID','CarName'],axis=1,inplace=True)
```

```python
df.head()
```

```python
df = pd.get_dummies(df,columns=['fueltype','aspiration','doornumber','carbody','drivewheel','enginelocation','enginetype','cylindernumber','fuelsystem'])
```

```python
df.head()
```

```python
from sklearn import preprocessing
```

```python
df.dtypes
```

```python
for i in df.columns:
    if i != 'price' and df[i].dtype != 'uint8':
        df[i] = preprocessing.scale(df[i].astype('float64'))
```

```python
df.head()
```

```python
df.to_csv('datasets/finaldata.csv',index=False)
```

```python
X = df.drop(['price','symboling'],axis=1)
Y = df['price']

```

```python
X.head()
```

```python
import statsmodels.api as sm
```

```python
X1 = X.copy() 
Y1 = Y.copy()
```

```python
X1 = np.array(X1)
X1 = sm.add_constant(X1)

model = sm.OLS(Y1,X1).fit()
model.summary()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3)
```

```python
lm = LinearRegression(fit_intercept=True).fit(x_train,y_train)
```

```python
print('training score: ',lm.score(x_train,y_train))
```

```python
y_pred = lm.predict(x_test)
```

```python
from sklearn.metrics import r2_score
```

```python
print('testing score: ',r2_score(y_test,y_pred))
```

```python
plt.figure(figsize = (12, 8))

plt.plot(y_pred, label='Predicted')
plt.plot(y_test.values, label='Actual')

plt.ylabel('price')

plt.legend()
plt.show()
```
